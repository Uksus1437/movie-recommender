```python
from sklearn.neighbors import NearestNeighbors
from scipy.sparse import load_npz
import pandas as pd

combined_matrix = load_npz("D:/vs_projects/movie-recommender/data/combined_matrix.npz")
preprocess_data = pd.read_csv('D:/vs_projects/movie-recommender/data/final_process_data.csv')
```

```python
def find_similar_movies(movie_index, nn_model, tfidf_matrix, n=30):
    query_vector = tfidf_matrix[movie_index]
    distances, indices = nn_model.kneighbors(query_vector, n_neighbors=n)
    return indices.flatten(), distances.flatten()
```

```python
nn_model = NearestNeighbors(n_neighbors=30, metric='cosine', algorithm='brute', n_jobs=-1)
nn_model.fit(combined_matrix)
```

```python
movie_id = 11
print(preprocess_data.iloc[movie_id][['title', 'genre', 'imdb_rating']])

similar_movies, scores = find_similar_movies(movie_id, nn_model, combined_matrix)

res_df = pd.DataFrame({'title': [], 'genre': [], 'imdb_rating': []})

for i in range(len(similar_movies)):
    if preprocess_data.iloc[similar_movies[i]]['title'] != preprocess_data.iloc[movie_id]['title']:
        res_df.loc[i] = preprocess_data.iloc[similar_movies[i]][['title', 'genre', 'imdb_rating']]
res_df.head(19)
```
